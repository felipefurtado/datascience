Agora que já transformamos os elementos que queríamos em booleans e conseguimos localizá-los e contá-los:

-------- vamos selecioná-los (**indexação de arrays**) utilizando a máscara booleana -------

SAINDO DO **EXEMPLO ALEATÓRIO** E VOLTANDO AO **DATAFRAME**

no exemplo nós queríamos os valores 1 do array

no dataframe vamos querer os valores da série(coluna) ------- id_counts ==2 -------

```python
dupe_mask = id_counts ==2
```

```python
dupe_mask
```

Os IDs agora são os índices e os booleans os valores da série id_counts

```python
id_counts.index[:]
```

**SELECIONE E ARMAZENE** os IDs duplicados(Trues) em uma variável

```python
dupe_ids = id_counts.index[dupe_mask]
```

Converta a variável para uma lista

```python
dupe_ids = list(dupe_ids)
```

```python
len(dupe_ids)
```

```python
dupe_ids
```

```python
# verificando os valores das características dos IDs duplicados
df.loc[df['ID'].isin(dupe_ids[0:3]),:]
```

Vamos encontrar as linhas que somente tem zeros nas colunas (exceto coluna ID)

VAMOS **CRIAR UMA MATRIZ BOOLEANA** COM O **MESMO TAMANHO DO DATAFRAME**

------> USANDO A CONDIÇÃO É == 0

```python
df_zero_mask = df == 0
```

```python
df_zero_mask
```

AGORA PRECISAMOS **INDEXAR A MATRIZ BOOLEANA -> df_zero_mask** USANDO A INDEXAÇÃO DE INTEIROS (.iloc)

```python
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
```

```python
feature_zero_mask
```

```python
sum(feature_zero_mask)
```

315 linhas tem 0 em todas as colunas

------ **APAGANDO** AS LINHAS COM ZEROS EM TODAS AS COLUNAS -> ELIMINANDO OS IDS DUPLICADOS --------

```python
df_clean_1 = df.loc[~feature_zero_mask,:].copy()
```

ASSIM **SELECIONAMOS** ATRAVÉS DO (.loc) TODAS AS **LINHAS QUE NÃO POSSUEM ZEROS EM TODAS AS COLUNAS**

Criando uma cópia ao invés de ver a saída do DataFrame original

```python
# verificamos a cópia

df_clean_1
```

```python
# agora verificamos se o número de linhas restantes é igual ao número de IDs exclusivos

# primeiro o número de linhas restantes

df_clean_1.shape
```

```python
# agora o número de IDs exclusivos

df_clean_1['ID'].nunique()
```

```python
# IDs duplicados exclídos com sucesso!
```
